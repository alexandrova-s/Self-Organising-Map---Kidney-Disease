```
from __future__ import print_function
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import io
from google.colab import files
```

Wczytujemy plik z analizowanym zbiorem danych do sieci neuronowej.

```
uploaded = files.upload()
```

```
df = pd.read_csv(io.BytesIO(uploaded['csv_result-chronic_kidney_disease_full.csv']), error_bad_lines=False, na_values='?', index_col = "class")
df
```

Zamieniamy dane nienumeryczne na wartoci logiczne 0 i 1.

```
for i in range(0, (len(df))):
  for j in [6,7,8,9,19,20,21,22,23,24]:
    if df.iloc[i, j] in {'normal', 'notpresent', 'no', 'good'}:
      df.iloc[i, j] = 0
    elif df.iloc[i, j] in {'abnormal', 'present', 'yes', 'poor'}:
      df.iloc[i, j] = 1 
    
```

Opisanie danych podstawowymi funkcjami statystycznymi.

Dzielimy zbir na dwa podzbiory: osoby chore i bez wykrytej choroby, by zobaczy jakie cechy o niej wiadcz.

```
ill = df.loc[['ckd']]
nill = df.loc[['notckd']]
```

```
ill.describe()
```

```
nill.describe()
```

Z danego opisu moemy odczta takie dane, jak np. rednia, czy odchylenie standardowe, co pozwala nam lepiej zapozna si ze zbiorem danych. 
Np. widzimy, e redni wiek pacjentw zdiagnozowanych na przewlek niedoczynno nerek to ok. 55 lat i natomiast pacjentw, u ktrych nie zdiagnozowano choroby - 46 lat, wic moemy powiza ten czynnik z ryzykiem zachorowania na przewlek niewydolno nerek.

Dane obrazujemy za pomoc histogramw dla kadej cechy, ktre pomog nam oceni warto cechy pod wzgldem badanego projektu.
Histogramy mona spodzi tylko dla danych liczbowych, wic nie zobrazuj nam wszystkich rozpatrywanych cech.
Porwnujemy histogramy kadej z cech dla pacjentw chorych i zdrowych.

```
for eleme in ["age", "al", "bgr", "bp", "bu", "hemo", "pcv", "pot", "rbcc", "sc", "sg", "sod", "su", "wbcc"]:  
  fig, axes = fig, axes = plt.subplots(nrows=1, ncols=2)
  ax0, ax1 = axes.flatten()

  ax0.hist(ill[eleme] , ec ='k', rwidth = .9, color='r')
  ax0.set_title(eleme + ' ckd')
  ax1.hist(nill[eleme] , ec ='k', rwidth = .9, color='g')
  ax1.set_title(eleme + ' notckd')


plt.show()
```

Jak wida, wiele histogramw daje nam bardzo przydatne informacje, poniewa jest w nich pewna dominujca warto, ktra moe mie bardzo due powizanie z wartoci ryzyka zachorowania na przewlek niewydolno nerek. 
Dominujc warto, ktra na pewno moe mie wpyw jest np. liczba upakowanie komrek w objtoci krwi, liczba krwinek biaych i czerwonych, czy te poziom glukozy we krwi.

Dane nienumeryczne przdstawimy w postaci procentowej zawartoci wartoci normalnej i niepoprawnej danego parametru w grupie osb chorych i zdrowych, dziki czemu zobaczymy, czy istnieje ich powizanie z wystpowaniem choroby.

```
print("Procenrowa zawarto wartoci niepoprawnej u osb chorych")
for param in ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'pe', 'ane', 'appet']:
  sum_normal = 0
  sum_abnormal = 0
  for el in ill[param]:
    if el == 0:
      sum_normal += 1
    elif el == 1:
      sum_abnormal += 1

  print(param, round(sum_abnormal/(sum_normal + sum_abnormal)*100, 3) , "%")
```

```
print("Procenrowa zawarto wartoci niepoprawnej u osb zdrowych")
for param in ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'pe', 'ane', 'appet']:
  sum_normal = 0
  sum_abnormal = 0
  for el in nill[param]:
    if el == 0:
      sum_normal += 1
    elif el == 1:
      sum_abnormal += 1

  print(param, round(sum_abnormal/(sum_normal + sum_abnormal)*100, 3) , "%")
```
