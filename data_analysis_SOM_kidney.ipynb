```
from __future__ import print_function
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import io
from google.colab import files
```

Wczytujemy plik z analizowanym zbiorem danych do sieci neuronowej.

```
uploaded = files.upload()
```

```
df = pd.read_csv(io.BytesIO(uploaded['csv_result-chronic_kidney_disease_full.csv']), error_bad_lines=False, na_values='?', index_col = "class")
df
```

Zamieniamy dane nienumeryczne na wartości logiczne 0 i 1.

```
for i in range(0, (len(df))):
  for j in [6,7,8,9,19,20,21,22,23,24]:
    if df.iloc[i, j] in {'normal', 'notpresent', 'no', 'good'}:
      df.iloc[i, j] = 0
    elif df.iloc[i, j] in {'abnormal', 'present', 'yes', 'poor'}:
      df.iloc[i, j] = 1 
    
```

Opisanie danych podstawowymi funkcjami statystycznymi.

Dzielimy zbiór na dwa podzbiory: osoby chore i bez wykrytej choroby, by zobaczyć jakie cechy o niej świadczą.

```
ill = df.loc[['ckd']]
nill = df.loc[['notckd']]
```

```
ill.describe()
```

```
nill.describe()
```

Z danego opisu możemy odcztać takie dane, jak np. średnia, czy odchylenie standardowe, co pozwala nam lepiej zapoznać się ze zbiorem danych. 
Np. widzimy, że średni wiek pacjentów zdiagnozowanych na przewlekłą niedoczynność nerek to ok. 55 lat i natomiast pacjentów, u których nie zdiagnozowano choroby - 46 lat, więc możemy powiązać ten czynnik z ryzykiem zachorowania na przewlekłą niewydolność nerek.

Dane obrazujemy za pomocą histogramów dla każdej cechy, które pomogą nam ocenić wartość cechy pod względem badanego projektu.
Histogramy można spożądzić tylko dla danych liczbowych, więc nie zobrazują nam wszystkich rozpatrywanych cech.
Porównujemy histogramy każdej z cech dla pacjentów chorych i zdrowych.

```
for eleme in ["age", "al", "bgr", "bp", "bu", "hemo", "pcv", "pot", "rbcc", "sc", "sg", "sod", "su", "wbcc"]:  
  fig, axes = fig, axes = plt.subplots(nrows=1, ncols=2)
  ax0, ax1 = axes.flatten()

  ax0.hist(ill[eleme] , ec ='k', rwidth = .9, color='r')
  ax0.set_title(eleme + ' ckd')
  ax1.hist(nill[eleme] , ec ='k', rwidth = .9, color='g')
  ax1.set_title(eleme + ' notckd')


plt.show()
```

Jak widać, wiele histogramów daje nam bardzo przydatne informacje, ponieważ jest w nich pewna dominująca wartość, która może mieć bardzo duże powiązanie z wartością ryzyka zachorowania na przewlekłą niewydolność nerek. 
Dominującą wartość, która na pewno może mieć wpływ jest np. liczba upakowanie komórek w objętości krwi, liczba krwinek białych i czerwonych, czy też poziom glukozy we krwi.

Dane nienumeryczne przdstawimy w postaci procentowej zawartości wartości normalnej i niepoprawnej danego parametru w grupie osób chorych i zdrowych, dzięki czemu zobaczymy, czy istnieje ich powiązanie z występowaniem choroby.

```
print("Procenrowa zawartość wartości niepoprawnej u osób chorych")
for param in ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'pe', 'ane', 'appet']:
  sum_normal = 0
  sum_abnormal = 0
  for el in ill[param]:
    if el == 0:
      sum_normal += 1
    elif el == 1:
      sum_abnormal += 1

  print(param, round(sum_abnormal/(sum_normal + sum_abnormal)*100, 3) , "%")
```

```
print("Procenrowa zawartość wartości niepoprawnej u osób zdrowych")
for param in ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'pe', 'ane', 'appet']:
  sum_normal = 0
  sum_abnormal = 0
  for el in nill[param]:
    if el == 0:
      sum_normal += 1
    elif el == 1:
      sum_abnormal += 1

  print(param, round(sum_abnormal/(sum_normal + sum_abnormal)*100, 3) , "%")
```
